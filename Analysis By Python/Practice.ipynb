```python
import numpy as np
from scipy.stats import trim_mean
```

```python
col=df1['ApplicantIncome']
Q1=col.quantile(0.25)
Q3=col.quantile(0.75)
IQR=Q3-Q1
lower_bound=Q1-1.5*IQR
upper_bound=Q3+1.5*IQR
Outliers=(col<lower_bound)|(col>upper_bound)

```

```python
trim_perc=sum(Outliers)/len(col)
trim_perc
trim_mean_result=trim_mean(df1['ApplicantIncome'],proportiontocut=trim_perc)
trim_mean_result
trim_mean_result1=trim_mean(df1['CoapplicantIncome'],proportiontocut=trim_perc)
trim_mean_result2=trim_mean(df1['LoanAmount'],proportiontocut=trim_perc)
df1['ApplicantIncome']=np.where(Outliers,trim_mean_result,col)
df1['CoapplicantIncome']=np.where(Outliers,trim_mean_result1,col)
df1['LoanAmount']=np.where(Outliers,trim_mean_result2,col)
```

```python
trim_mean_result2=trim_mean(df1['LoanAmount'],proportiontocut=trim_perc)
df1['LoanAmount'].fillna(trim_mean_result2,inplace=True)
```
