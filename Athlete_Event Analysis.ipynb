```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv("athlete_events.csv")
df.head()
```

```python
winners = df.dropna(subset=['Medal'])
winners.head()
```

```python
winners_2016 = winners[(winners.Year==2016)]
winners_2016.head()
```

```python
#top 5 sports
winners_2016.Sport.value_counts()
```

Top 5 sports are Athletics, Swimming, Rowing, Football, Hockey

```python
top_5_sports = ['Athletics','Swimming','Rowing','Football','Hockey']
```

```python
top_sports_winners_2016 = winners_2016[(winners_2016.Sport.isin(top_5_sports))]
top_sports_winners_2016.head()
```

#1: 2016 Top Sport Winners

```python
ex1 = sns.catplot(x='Sport',data=top_sports_winners_2016,kind='count',aspect=1.5)
```

#2: Age distribution of athletes in top 5 sports in 2016

```python
ex2 =sns.displot(top_sports_winners_2016.Age,kde=False)
```

#3: Country-wise medals won for top 5 countries

```python
ex3 = sns.catplot(x='Team',data=top_sports_winners_2016,kind='count',aspect=1.5,order=top_sports_winners_2016['Team'].value_counts().index)
ex3.set_xticklabels(fontsize=6,rotation=90)
```

#4: Average Weight of Winners based on Gender for top 5 sports

```python
sns.set(style="whitegrid")
sns.barplot(data=top_sports_winners_2016,x='Sport',y='Weight',estimator="mean",hue="Sex")
```
